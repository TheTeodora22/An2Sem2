```python
from __future__ import annotations
from typing import Any
```

Clasa corespunzatoare nodurilor din arborele de cautare.

```python
class NodArbore:
    #constructorul
    def __init__(self, informatie: Any, parinte: NodArbore = None):
        self.informatie = informatie
        self.parinte = parinte

    def drumRadacina(self) -> list [NodArbore]:
        nod=self
        lDrum=[]
        while nod:
            lDrum.append(nod)
            nod=nod.parinte
        return lDrum[::-1]

    def inDrum(self, infonod : Any) -> bool :
        nod=self
        while nod:
            if nod.informatie==infonod:
                return True
            nod=nod.parinte
        return False


    def __str__(self) -> str :
        return str(self.informatie)

    #c (a->b->c)
    def __repr__(self) -> str :
        return f"{self.informatie}, ({'->'.join([str(x) for x in self.drumRadacina()])})"
```

Clasa corespunzatoare grafului problemei

```python
class Graf:
    def __init__(self, matr: list[list[int]], start: Any , scopuri: list[Any]):
        self.matr=matr
        self.start=start
        self.scopuri=scopuri

    def scop(self, informatieNod: Any) -> bool:
        return informatieNod in self.scopuri

    def succesori(self, nod: NodArbore) -> list[NodArbore]:
        lSuccesori=[]
        for infoSuccesor in range(len(self.matr)):
            conditieMuchie = (self.matr[nod.informatie][infoSuccesor] == 1)
            conditieNotInDrum = not nod.inDrum(infoSuccesor)
            if conditieMuchie and conditieNotInDrum:
                lSuccesori.append(NodArbore(infoSuccesor, nod))
        return lSuccesori
```

<b>DepthFirst</b>
<p>Pasi:</p>
<ul>
    <li>Se pune nodul start într-o stivă.</li>
    <li>Repetitiv:</li>
        <ul>
            <li>Dacă vârful stivei este nod scop, afișăm o soluție. Dacă am ajuns la numărul de soluții dorit, ne oprim.</li>
            <li>Pentru vârful stivei dacă nu i s-au generat succesorii deja, adăugăm succesorii săi nevizitați pe drumul curent din stivă și verificăm dacă e nod scop, caz în care se returnează o soluție. Dacă am ajuns la numărul de soluții dorit, ne oprim.</li>
            <li>Dacă vârful stivei nu are succesori sau au fost deja generați și procesați, este eliminat din stivă.</li>
        </ul>

<p>Definiti functia depth_first care primește ca parametri un obiect de tip Graf  și un număr de soluții de afișat.</p>
